```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#setup

Nmu=50;    
Nw=500;

mumin=.001;
mumax=10;
dmu=(mumax-mumin)/Nmu

wmin=0.01;
wmax=6.;
dw=(wmax-wmin)/Nw;

Frequencies = np.array([wmin+dw*w for w in range(Nw)])

mockrho = [0.1]*Nw

for j in range(Nw):
    mockrho[j]=Frequencies[j]/((Frequencies[j]**2-3**2)**2+2**2)

#Each row of the kernel is "integrals  of the response kernels", eq. 18.6.5
Kernel=np.array([[  (1./2. if w == 0 or w== Nw-1 else 1.) * 
                    dw*2*(wmin+dw*w)/((wmin+dw*w)**2+(mumin+dmu*mu)**2)
                  for w in range(Nw) ] for mu in range(1,Nmu+1)]);


DataIdeal = np.dot(Kernel,mockrho)

print(DataIdeal.shape)
print(Kernel.shape)
plt.plot(mockrho)
```

```python
NoisyData = [0]*Nmu

#standard deviation scaling
K=1e-3
Nmeas =1000

for i in range(Nmu):
    NoisyData[i]=(np.random.normal(loc=DataIdeal[i],scale=DataIdeal[i]*K*np.sqrt(Nmeas),size=Nmeas))
    
#print(NoisyData)
```

```python
DataAverage = [0]*len(NoisyData)
Variance= [0]*len(NoisyData)

for j in range(len(NoisyData)):
    DataAverage[j]=np.mean(NoisyData[j])
    Variance[j] = np.var(NoisyData[j])
    
print(DataAverage)
print(Variance)
```

```python
plt.yscale('log')
plt.xscale('log')
plt.plot(DataAverage)
#plt.plot(DataIdeal)
```

```python
AutoCorrelationCorr=1.; 

#standard covariance expression, why (Nmeas-1.)*Nmeas?
DataCovarianceMatrix=AutoCorrelationCorr*np.array([[ sum( np.array([ ( NoisyData[i][n]-DataAverage[i] )*
                                     ( NoisyData[j][n]-DataAverage[j] )/((Nmeas-1.)*Nmeas) for n in range(Nmeas)]) )
                  for i in range(Nmu) ] for j in range(Nmu)]);
print(DataCovarianceMatrix.shape)
print(DataCovarianceMatrix)
```

```python

RM=np.array([[  2*(wmin+dw*w)/((wmin+dw*w)**2+(mumin+dmu*mu)**2)
                  for w in range(Nw) ] for mu in range(1,Nmu+1)]);

R = np.array([sum(RM[i,j]*(0.5 if j==0 or j==Nw-1 else 1.) for j in range(Nw)) for i in range(Nmu)])*dw
S = DataCovarianceMatrix
```

```python
def W(wmin,wmax,w,i,j):
    result = 0
    if(i==j):
        for x in (-wmin,wmax):
            result += 4*(x + (-2*w*i*i-w*w*x+i*i*x)/(2*(i*i+x*x)) + (w**3-3*i*i)*np.arctan(x/i)/(2*i) - np.log(i*i+x*x))
    else:
        for x in (-wmin,wmax):
            result += 4*(i*i*x - j*j*x + i*(w**2-i**2)*np.arctan(x/i) + (-w**2*j+j**3)*np.arctan(x/j) - w*i**2*np.log(i**2+x**2) + w*j**2*np.log(j**2+x**2))/(i**2-j**2)
    
    return(result)
```

```python
W = np.empty((Nmu,Nmu))
u = np.empty(Nw)
for x in range(Nw):
    
    print(x)
   
    wcurrent = wmin+dw*x
    
    #Computing spread matrix
    for i in range(Nmu):
        for j in range(Nmu):
            W[i,j]=0
            if(i > j):
                W[i,j] = W[j,i]
            else:
                for w in range(Nw):
                    dwc = (0.5 if w == 0 or w== Nw-1 else 1.)*dw
                    W[i,j] += (Frequencies[w]-wcurrent)**2*RM[i,w]*RM[j,w]*dwc
            
    winv = np.linalg.solve((W+1*S),R)
    
    u[x] = (DataAverage@winv)/(R@winv)
    
#print(DCDataAvg.shape, winv.shape, R.shape)
plt.plot(u)
```

```python
Wa = np.empty((Nmu,Nmu))
u = np.empty(Nw)
for x in range(Nw):
    
    print(x)
   
    wcurrent = wmin+dw*x
    
    #Computing spread matrix
    for i in range(Nmu):
        imu = mumin+dmu*i
        for j in range(Nmu):
            jmu = mumin+dmu*j
            Wa[i,j]=0
            if(i > j):
                Wa[i,j] = Wa[j,i]
            else:
                Wa[i,j] = W(wmin, wmax, wcurrent, imu, jmu)
            
    winv = np.linalg.solve((Wa+1*S),R)
    
    u[x] = (DataAverage@winv)/(R@winv)
    
#print(DCDataAvg.shape, winv.shape, R.shape)
plt.plot(u)
```
